```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("customer_churn_dataset.csv")
```

```python
# Line Chart
plt.figure(figsize=(10,5))
plt.plot(df.index[:50], df['MonthlyCharges'][:50], marker='o')
plt.title("Monthly Charges Trend (First 50 Customers)")
plt.xlabel("Customer Index")
plt.ylabel("Monthly Charges")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Bar chart
plt.figure(figsize=(6,4))
df['IsActive'].value_counts().plot(kind='bar')
plt.title("Active vs Churned Customers")
plt.xlabel("Status (1 = Active, 0 = Churned)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
#Histogram
plt.figure(figsize=(8,5))
plt.hist(df['Age'], bins=10, edgecolor='black')
plt.title("Age Distribution of Customers")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
#Box plot 
plt.figure(figsize=(8,5))
sns.boxplot(x='ContractType', y='MonthlyCharges', data=df)
plt.title("Monthly Charges by Contract Type")
plt.tight_layout()
plt.show()
```

```python
#Scatter plot
plt.figure(figsize=(8,5))
plt.scatter(df['MonthlyCharges'], df['TotalCharges'], alpha=0.7)
plt.title("Monthly Charges vs Total Charges")
plt.xlabel("Monthly Charges")
plt.ylabel("Total Charges")
plt.tight_layout()
plt.show()
```

```python
#Heat map
plt.figure(figsize=(8,6))
sns.heatmap(df[['Age','TenureMonths','MonthlyCharges','TotalCharges','SupportTickets','IsActive']].corr(),
            annot=True, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
#Pie chart
payment_counts = df['PaymentMethod'].value_counts()

plt.figure(figsize=(7,7))
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Payment Method Distribution")
plt.tight_layout()
plt.show()
```

```python
#Area Chart
area_data = df.groupby('TenureMonths')['MonthlyCharges'].mean().reset_index()

plt.figure(figsize=(10,5))
plt.fill_between(area_data['TenureMonths'], area_data['MonthlyCharges'], alpha=0.5)
plt.plot(area_data['TenureMonths'], area_data['MonthlyCharges'])
plt.title("Average Monthly Charges by Tenure (Area Chart)")
plt.xlabel("Tenure (Months)")
plt.ylabel("Average Monthly Charges")
plt.tight_layout()
plt.show()
```
